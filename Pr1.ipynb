```python
#linear encoding

df['gender_encoded'] = df['gender'].cat.codes
df['lunch_encoded'] = df['lunch'].cat.codes
df['test_prep_encoded'] = df['test preparation course'].cat.codes
df['parent_edu_encoded'] = df['parental level of education'].cat.codes
```

```python
#one hot encoding

df['gender'] = df['gender'].astype('category')
df['lunch'] = df['lunch'].astype('category')
df['test preparation course'] = df['test preparation course'].astype('category')
df['parental level of education'] = df['parental level of education'].astype('category')
```

```python
print(df.isnull().sum())
```

```python
df['math score'].fillna(df['math score'].median(), inplace=True)
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
df['lunch'] = df['lunch'].replace({'standard ': 'standard', 'free':'free/reduced'})
```

```python
df.head(10)
```

```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['math score'] = df['math score'].clip(lower, upper)
```
